### Exploratory Data Analysis

Exploratory Data Analysis(EDA) is essentially a type of storytelling for statisticians. It is a way of visualizing, summarizing and interpreting the data that is hidden in rows and column format. Once EDA is complete and insights are drawn, its feature can be used for supervised and unsupervised machine learning modelling.

### SampleSuperstore 

The Super Store dataset contains data on order details of customers for orders of a superstore in the US. This includes the state, region, order date, shipping date, product ordered etc.

```python
# Importing the required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Importing the dataset
df = pd.read_csv('SampleSuperstore.csv')
df.head()
```

##### Checking the dataset for duplicate values and dropping that element

```python
df.duplicated().sum()
```

```python
df.drop_duplicates()
```

##### Removing the unrequired variable from the dataset

```python
df = df.drop(columns = 'Postal Code', axis = 1)
```

```python
# Checking the shape to confirm 
df.shape
```

##### Visualizing the dataset

```python
print(df['State'].value_counts())
plt.figure(figsize = (15, 8))
sns.countplot(x = df['State'])
plt.xticks(rotation = 90)
plt.show()
```

```python
print(df['Sub-Category'].value_counts())
plt.figure(figsize = (12, 6))
sns.countplot(x = df['Sub-Category'])
plt.xticks(rotation = 90)
plt.show()
```

##### Plotting the heatmap to check the corealtion between the variables

```python
fig, axes = plt.subplots(1, 1, figsize = (9, 6))
sns.heatmap(df.corr(), annot = True)
plt.show()
```

##### Countplots on different variables

```python
sns.countplot(x = df['Segment'])
```

```python
sns.countplot(x = df['Region'])
```

##### Plotting the profits with respect to sub categories

```python
plt.figure(figsize = (40, 25))
sns.barplot(x = 'Sub-Category', y = 'Profit', data = df)
plt.xticks(rotation = 90, fontsize = 25)
```

##### Checking the profit on different discounts

```python
plt.figure(figsize = (10, 4))
sns.lineplot('Discount', 'Profit', data = df, color = 'blue', label = 'Discount')
plt.legend()
```

##### Plotting a Histogram for all the continous variables

```python
df.hist(bins = 50, figsize = (30, 20))
plt.show()
```

##### Pairplot based on Sub Categories

```python
figsize = (20, 15)
sns.pairplot(df, hue = 'Sub-Category')
```

##### Grouping the sum of sales of every sub-category to every state of region

```python
df_grp = pd.DataFrame(df.groupby(['Ship Mode', 'Segment', 'Category', 'Sub-Category', 'State', 'Region'])['Quantity', 'Discount', 'Sales', 'Profit'].sum().reset_index())
df_grp
```

##### Calculating the sum, mean, max, min, median, standard deviation, variance of Profit with respect to each state

```python
df.groupby('State')['Profit'].agg(['count', 'sum', 'mean', 'max', 'min', 'median', 'std', 'var'])
```
